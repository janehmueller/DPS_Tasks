```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.python import keras
from tensorflow.python.keras.models import load_model
```

```python
model_path = "C:/repo/data-knoller/data-prep/dataprep-simple/src/main/resources/model_emb3_epochs2.hdf5"
model = load_model(model_path)
```

```python
emb = model.get_layer("embedding")
weights = np.array(emb.get_weights())
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(weights[0, :, 0], weights[0, :, 1], weights[0, :, 2])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
