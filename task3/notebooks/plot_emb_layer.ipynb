```python
import json
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.python import keras
from tensorflow.python.keras.models import load_model
```

```python
model_path = "C:/repo/data-knoller/data-prep/dataprep-simple/src/main/resources/second/model_emb3_epochs5.hdf5"
model = load_model(model_path)
```

```python
with open("C:/repo/data-knoller/data-prep/dataprep-simple/src/main/resources/second/vocab.json", "rb") as file:
    voc_dict = json.loads(file.read())
```

```python
inv_voc_dict = {value: key for key, value in voc_dict.items()}
```

```python
emb = model.get_layer("embedding_1")
weights = np.array(emb.get_weights())
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111, projection='3d')

for i in range(weights.shape[1]):
    x = weights[0, i, 0]
    y = weights[0, i, 1]
    z = weights[0, i, 2]
    print(f"{x}, {y}, {z}, \"{inv_voc_dict[i]}\"")
    ax.scatter(x, y, z)
    ax.text(x, y, z, inv_voc_dict[i], size=15)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.tight_layout()
plt.show()
```
